```python
#Importing the libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Seaborn has some built-in dataset

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
#visualize the data
sns.relplot(x="total_bill",y="tip",col="time",hue="smoker",style="smoker",size="size",data=tips);

```

```python
# house price dataset
# house price dataset
from sklearn.datasets import load_iris
house_iris = load_iris()

house = pd.DataFrame(house_iris.data, columns=house_iris.feature_names)
house['PRICE'] = house_iris.target

```

```python
print(house_iris)
```

```python
house.head()
```

Distribution Plot

```python
sns.distplot(house['PRICE'])
```

Correlation
positive correlation
negative correlation

heat map

```python
correlation=house.corr()
```

```python
#constructing a heat map
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
