Matplotlib

*   Useful for making Plots




```python
#importing Matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plot
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title,x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin value')
plt.title('sine Wave')
plt.show()
```

```python
#parabola
x= np.linspace(-5,5,100)
y=x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('parabola')
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x , y , ' g.')
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
x= np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

**Bar plot**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```
