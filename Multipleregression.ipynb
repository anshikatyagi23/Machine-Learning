Multiple Linear Regression Car Price Prediction

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/CarPrice_Assignment(1).xls')
```

```python
from sklearn.linear_model import LinearRegression
x = ['carlength', 'carwidth', 'curbweight']
y = ['price']

X = df[x]
Y = df[y]
```

```python
x="carwidth,carlength,curbweight"
y="price"
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
prediction_price=model.predict([[400,190,40]])
print(prediction_price)
```
